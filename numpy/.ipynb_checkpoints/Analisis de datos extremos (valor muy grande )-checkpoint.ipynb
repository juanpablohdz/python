```python
import pandas as pd
import numpy as np 
df= pd.read_csv("C:/Users/polla/Desktop/curso python esecencial ciencia de datos/numpy/datos1.csv",nrows=100000)
```

```python
df.head()
```

```python
x= df["SquareMeters"].dropna() # vamos a seleccionar 10000  de la bd y luego una columna
```

```python
Q1= np.percentile(x,25)# medida estadistica de los cuartiles
Q3= np.percentile(x,75)
rangointer=Q3-Q1  # rango intercuantilico diferencia entre el 3er cuartil y el 1ero
```

```python
umbralsuperior =  Q3 + 1.5*rangointer  #el primer umbral superior  por el 1.5 del rango inter
umbralinferior =  Q1 -1.5*rangointer # el umbral inferior  - el rango inter 
```

```python
umbralsuperior #un valor mayor que 450 va hacer un aoutlayer por cada distribuicion
```

```python
umbralinferior # lo mismo por debajo de nuestro umbral inferior 
```

```python
np.mean(x>umbralsuperior) # la media del umbral superior  y dice que el 7% esta por encima
```

```python
np.mean(x<umbralinferior) #no llega al 2% no son simetricos entonces nos da un umbral o un sesgo
```

```python
from sklearn.covariance import EllipticEnvelope #estudiar varias variables al mismo tiempo 
```

```python
outliers=EllipticEnvelope(contamination=.01)#crear modelo que selecciona el 1 % de datos  que considere que estan muy alejados
#de nuestros datos mas centrados
```

```python
var_list=["Price" ,"Bathrooms","Bedrooms","ParkingSlots","SquareMeters"]#para puros datos cualitativos con numero 
```

```python
x= np.array(df.loc[:,var_list].dropna()) #
```

```python
outliers.fit(x)
```

```python
pred = outliers.predict(x)
```

```python
pred # lo que nos interesa son los -1
```

```python
elips_outliers=np.where(pred==-1)[0]
```

```python
elips_outliers 
#conseguimos lista de todos los index cuyos casos que estan muy alejados de los valores centrales de todas las variables estudiadas
```
