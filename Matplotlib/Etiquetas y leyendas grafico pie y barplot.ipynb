```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
df=pd.read_csv("C:/Users/polla/Desktop/curso python esecencial ciencia de datos/Matplotlib/datos1.csv",nrows=10000)
data=np.unique(df.Bathrooms,return_counts=True)
labs=["baños1","baños2","baños3","baños4"]
data
```

```python
plt.pie(x = data[1],
       labels =labs,
       radius =.7,
    colors=sns.color_palette("hls",4), #hace colores aleatorios
        explode=(0,0.4,0,0),  #separa cacho de la grafica de pastel 
       startangle = 90,
       autopct="%9.1f%%", # porcentaje de cada 1
       labeldistance=1.1)# distancia de nombre)
plt.legend(loc="lower right",labels=labs) # pone tabla poner algo  sin sentido en loc para ver que parametros se pueden usar 

plt.show()
```

```python
plt=sns.barplot(x=labs,y=data[1])
plt.set(xlabel="numero de baños por casa",ylabel="catidades de baños")
```
