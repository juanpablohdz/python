```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import time 
df=pd.read_csv("C:/Users/polla/Desktop/curso python esecencial ciencia de datos/Matplotlib/datos1.csv")
df2=df[df["Estado"].isin(["Yucatán ","Querétaro ","Ciudad de México"])]
df=df.head(500000)
```

```python
df.head()
```

```python
times=[] #lista vacioa
for i in np.arange(len(df)):#iteracion
    times.append(datetime.datetime(year=2020, month = df.loc[i,"Bedrooms"],day=df.loc[i,"Bathrooms"]))
    #mont"meses" day ' dias '
```

```python
times[100] #muetsra los datos generados
```

```python
df["Time"]=times #creamos una base de datos para el timepo llamada Time 
```

```python
#base de datos unida  con la variable Time  puede ser ora de salida y ORA DE LLEGADA y se aplicara la media
data = df.groupby(by=["Time"],as_index=False)["ParkingSlots","Bathrooms"].mean()
data.head()
```

```python
sns.lineplot(data["Time"],data ["ParkingSlots"]) #grafica por tiempo 
```

```python
data = df.groupby(by=["Time"])["ParkingSlots","Bathrooms"].mean() #ejemplo del parametro sin forzar el index
data.head()
```

```python
sns.lineplot(data=data) #muestra la serie temporal 
```

```python
times=[] #lista vacioa
for i in df2.index:#indexado
    times.append(datetime.datetime(year=2020, month = df2.loc[i,"Bedrooms"],day=df2.loc[i,"Bathrooms"]))
    #mont"meses" day ' dias '
df2["Time"]=times
```

```python
#ajustar margenes 
sns.set(rc={'figure.figsize':(15,15)}) #tamaño para la serie temporal 
sns.lineplot(x="Time", y="ParkingSlots",hue="Bathrooms",data =df2)#serie tenmporal 
#hue es un dato categorico  puede ser origen donde parte las horas del de y y x
```
