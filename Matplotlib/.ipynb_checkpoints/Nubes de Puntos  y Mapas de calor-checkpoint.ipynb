```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/polla/Desktop/curso python esecencial ciencia de datos/Matplotlib/datos1.csv")
df.dropna(inplace=True,subset=["Price","Bathrooms","ParkingSlots","Bedrooms"])
sns.set(rc={'figure.figsize':(15,10)}) #ajusta tamano de grafico 
```

```python
df.head(1000)
```

```python
df2=df[df["Estado"].isin(["Yucatán","Ciudad de México","Querétaro"])].sample(frac=1).head(1000)
sns.jointplot(df2["Price"],df2["SquareMeters"]) #muestra de puntos y que tan correlacionados estan los datos
```

```python
df3=df2[np.abs(df2["Price"])<.5]
df3=df2[np.abs(df2["SquareMeters"])<450]
```

```python
sns.jointplot(df2["SquareMeters"],df2["Price"],kind="hex")
```

```python
sns.jointplot(df3["SquareMeters"],df3["Price"],kind="hex")
```

```python
sns.jointplot(df3["Price"],df3["SquareMeters"],kind="kde") # densidad  y curbas de nivel
```

```python
gb_df=pd.DataFrame(df2.groupby(["Estado","Municipio"],as_index=False)["Price"].mean())
gb_df.head() # se ajusta con el primer filtro
```

```python
data=gb_df.pivot("Municipio","Estado","Price")
data
```

```python
sns.set(rc={'figure.figsize':(15,8)}) #ajuste de tamamo
sns.heatmap(data=data,annot=True,linewidths=.5) #muestra las casas caras dependiendo del  color 
```
