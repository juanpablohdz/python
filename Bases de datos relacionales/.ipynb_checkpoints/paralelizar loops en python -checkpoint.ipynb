```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed  # libreria para paralelizar  usar numpy que dispone

df= pd.read_csv("C:/Users/polla/Desktop/curso python esecencial ciencia de datos/Bases de datos relacionales/datos1.csv",nrows=100000)
```

```python
df.head()
```

```python
df.sub = df[['Price','Bathrooms','Bedrooms','ParkingSlots','SquareMeters','GpsLongitude','GpsLatitude']]
df.sub.head(10)
```

```python
#calcular o buscar  el retaso maximo 
def retraso_maximo(fila):
    if not np.isnan(fila).any ():
        names=['Price','Bathrooms','Bedrooms','ParkingSlots','SquareMeters','GpsLongitude','GpsLatitude'] #busca nombres de la fila
        return names [fila.index (max(fila))] #tomara un valor maximo  
    else:
        return "None"
```

```python
#bucle 
results = []
for fila in df.sub.values.tolist(): 
    results.append(retraso_maximo (fila))
```

```python
results
```

```python
#paralelisa  el proceso 
results= Parallel(n_jobs=2,backend="multiprocessing")(
            map (delayed (retraso_maximo),df.sub.values.tolist()))
```
