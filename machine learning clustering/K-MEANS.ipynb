**algoritmo k-means busca separar los datos de en un numero predefinido asegurando que tengan la misma varianza  al centride mas cercano**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  sklearn.cluster import KMeans 
df=pd.read_csv("C:/Users/polla/Desktop/curso python esecencial ciencia de datos/machine learning/datos1.csv",nrows=1e5)
newdf= df[["Price","SquareMeters"]].dropna()
```

```python
df.head()
```

```python
newdf= df[["Price","Bathrooms","Bedrooms","ParkingSlots","SquareMeters","GpsLongitude","GpsLatitude"]].dropna()
```

**se repite  la newdf  para  ver los datos  de mas de 3 variables arriba al final**

```python
kmeans=KMeans (n_clusters=5,random_state =0,n_jobs=-1).fit(newdf)
#parametro de clusters  4   puede ser 100  random el mismo 0  n-jobs acelerara el proceso  ajustar el modelo .fit
#probar con 10
kmeans.labels_
```

```python
np.unique (kmeans.labels_,return_counts=True) # el  muy dispar el primero esta muy cargado y el 1 esta muy bajo
```

```python
import matplotlib.pyplot as plt
plt.scatter (newdf["Price"],newdf["SquareMeters"],c =kmeans.labels_)
plt.show()
```

**muestra 3 grupos muy cargados y muy proximos entre ellos  e 2 grupos con mucha dispersion**

```python
kmeans.cluster_centers_ #centroides de cada grupo identificar como x y y
```
