```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering #  libreria para cluster separar de manera euclidiana (distancia de cada cluster ) 
import matplotlib.pyplot as plt 
df=pd.read_csv ("C:/Users/polla/Desktop/curso python esecencial ciencia de datos/machine learning/datos1.csv",nrows=1e4)
newdf=df[["Price","SquareMeters"]].dropna()
```

```python
df.head()
```

```python
newdf=df[["Price","Bathrooms","Bedrooms","ParkingSlots","SquareMeters"]] # ajustar con mas variables
```

```python
#ajuste de modelo  a 5 cluster 
clstr=AgglomerativeClustering(n_clusters=20)
clstr.fit(newdf)
#distancia recorrida  
#euclidean/l1/l2/manhattan/cosine/precomputed
#affinity="L1",linkage ="complete "
```

```python
import matplotlib.pyplot as plt
plt.scatter(newdf["Price"],newdf["SquareMeters"],c=clstr.fit_predict(newdf))
plt.show()
```
